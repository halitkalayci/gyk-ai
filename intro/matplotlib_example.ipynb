### Matplotlib

Veri görselleştirme kütüphanesidir. Temel olarak 2D grafikler çizmek için kullanılır. (Bazı ek pluginlerle 3D grafikler oluşturabilir.)

```python
import matplotlib.pyplot as plt
```

```python
x = [ 1,2,3,4,5 ]
y = [ 10,20,30,40,50 ]
```

```python
x,y
```

### Grafik 1 İsterleri ( Tekil Çizgi Grafiği )

- Çizgi rengini yeşil yapınız.
- X ekseni değerleri 0'dan başlayacak şekilde düzenleyiniz. (Arraydeki ilk değer 0 olmasın.)

```python
plt.plot(x,y, marker="s", linestyle="dotted", color="r", label="Satışlar1")
plt.xlabel("Günler")
plt.ylabel("Satışlar")
plt.title("Günlük Satışlar")
plt.legend()
plt.grid(True)
plt.show() # Jupyter notebook dışında çalışıyor isek, bunu yazmalıyız.
```

## Çoklu Grafik

İsterler:

- Yeni bir grafik tan,cot grafiğini çizelim.

```python
import numpy as np
```

```python
# 0 ile 10 arasında 100 adet araları eşit sayı oluştur
x = np.linspace(0, 10, 100)
x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)", color="red")
plt.plot(x, y2, label="cos(x)", color="blue")

plt.legend()
plt.xlabel("X Değeri")
plt.ylabel("Fonksiyon Değeri (sin,cos)")
plt.grid(True)
```

## Bar Grafiği

İster yok

```python
categories = [ "A","B","C","D","E","F" ]
values = [ 5, 10 ,20 ,40, 50, 30 ]
```

```python
plt.bar(categories, values, color = ["red","blue","green","purple","orange","black"])
plt.xlabel("Kategoriler")
plt.ylabel("Değerler")
plt.title("Kategori bazlı satışlar")
#plt.grid(True)
```

## Histogram (Veri Dağılımı)

```python
data = np.random.rand(1000)
```

```python
plt.hist(data, bins=30, color="skyblue", edgecolor="black")
plt.xlabel("Değer")
plt.ylabel("Frekans")
```
